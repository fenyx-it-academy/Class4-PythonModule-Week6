```python
class Rectangle:
    def __init__(self,length,width):   
        self.length=length
        self.width=width
    def findarea(self):
        return self.length*self.width
    def findperemetir(self):
        return 2*(self.length+self.width)
length=int(input("Enter length:"))
width=int(input("Enter width:"))
r1=Rectangle(length,width)
print("Area of rectangle :",r1.findarea())
print("peremetir of rentangle:",r1.findperemetir())


class Triangle:
    def __init__(self,length1,length2,length3):
        self.length1=length1
        self.length2=length2
        self.length3=length3
    def findperemetir(self):
        return (self.length1+self.length2+self.length3)
    def findarea(self):
        a=(self.length1+self.length2+self.length3)/2
        return(a*(a-self.length1)*(a-self.length2)*(a-self.length3))**0.5
length1=float(input("Enter the 1.length:"))
length2=float(input("Enter 2.length :"))
length3=float(input("Enter 3.length3 :"))
u=Triangle(length1,length2,length3)
print ("The Area of this Triangle is:",u.findarea())
print("The Perimeter of this Triangle is:",u.findperemetir())


class Square(Rectangle):
    def __init__(self,side):
        self.side=side
    def findarea(self):
        return self.side*self.side
    def findper(self):
        return 4*self.side
side=float(input("Enter side:"))
s=Square(side)
print ("Area of Square:",s.findarea())
print("Per of square:",s.findper())

class Cub(Square):
    def __init__(self,length):
        self.length=length
    def findarea(self):
        return 6*self.length*self.length
    def findvalume(self):
        return self.length*self.length*self.length
length=float(input("Enter length:"))
c=Cub(length)
print("Area of cub:",c.findarea())
print("Valume of cub:",c.findvalume())

class Pyramid (Triangle,Square):
    def __init__(self,side,length1,length2):
        super().__init__(side,length1,length2)
        self.side=side
        self.length1=length1
        self.length2=length2
    def findvalume(self):
        return ((self.length1**2-(self.side**2)/2)**0.5)*((1/3)*(self.side**2))
    
p=Pyramid(side,length1,length2)
print ("Volume of Pyramid:",int(p.findvalume()))
            
        
     
        
```
